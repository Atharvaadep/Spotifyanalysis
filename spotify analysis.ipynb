```python
corr_df= df_tracks.drop(['key', 'mode', 'explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True,fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')#
heatmap.set_title('CHMBV')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete = True, aspect=2, height=5, kind ='hist').set(title ='Songs per year')
```

```python
total_dr = df_tracks.duration_ms
fig_dms = (18,7)
fig , ax = plt.subplots(figsize = fig_dms)
fig = sns.barplot(x = years, y = total_dr , ax = ax, errwidth = False).set(title = 'Years vs Duration') 
plt.xticks(rotation = 90)
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10, 5))
Famous = df_genre.sort_values('popularity', ascending= False).head(10)
sns.barplot(x = 'popularity', y = 'genre' , data = Famous).set(title = 'Popularity vs Followers') 
```

```python
df_genre['genre'].unique()
```

```python
genre_pie = ['Movie', 'R&B', 'A Capella', 'Alternative', 'Country', 'Dance',
       'Electronic', 'Anime', 'Folk', 'Blues', 'Opera', 'Hip-Hop',
       "Children's Music", 'Children’s Music', 'Rap', 'Indie',
       'Classical', 'Pop', 'Reggae', 'Reggaeton', 'Jazz', 'Rock', 'Ska',
       'Comedy', 'Soul', 'Soundtrack', 'World']
data = df_genre['genre'].value_counts()

fig, ax = plt.subplots()
plt.pie(data,labels = genre_pie, autopct ='%.0f%%')
ax.set_title(' Total genre')
plt.show()
```
